```python
def f(x):
    if (isinstance(x, list)):
        res = 0
        for i in x:
            res += f(i)
        return res
    else:
        if x == 1:
            return None
        return 1 / (1 - x)


def geom_reihe(x, N):
    res = 0
    for i in range(0, N + 1):
        res += x ** i
    return res
```

```python
# Plotten Sie beide Funktionsgraphen uber dem Intervall I = [−1, 0.9] mit 100 Stutzstellen ¨
# und N = 4. Was stellen Sie fest? Untersuchen Sie anhand der Graphiken die jeweiligen
# Definitionsbereiche.

import numpy as np

I = range(-100, 90, 1)
N = 4

xpoints = []
graph_fx = []
graph_geom = []
for _x in I:
    x = _x / 100
    xpoints.append(x)
    graph_fx.append(f(x))
    graph_geom.append(geom_reihe(x, N))

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

plot_fx = ax.plot(xpoints, np.array(graph_fx))
plot_geo = ax.plot(xpoints, np.array(graph_geom))
ax.legend(["f(x) = 1 / (1 - x)", "f(x) = SUM(i=0, ^N, [x**i])"])

plt.show()
```
